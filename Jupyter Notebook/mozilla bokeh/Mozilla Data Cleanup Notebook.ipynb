```python
### Remove Null Values from questions and country column
new_dataframe = new_dataframe[new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:'].notnull()]
new_dataframe = new_dataframe[new_dataframe['Country'].notnull()]
new_dataframe = new_dataframe[new_dataframe['What is your biggest fear as we move towards a more connected future?'].notnull()]
new_dataframe = new_dataframe[new_dataframe['What are you most excited about as we move toward a more digitally connected future?'].notnull()]
new_dataframe = new_dataframe[new_dataframe['I consider myself:'].notnull()]
new_dataframe
```

```python
#Removing Other () User specified responses 
new_dataframe = new_dataframe[new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:'] != 'Other (please specify)' ]
new_dataframe = new_dataframe[new_dataframe['What is your biggest fear as we move towards a more connected future?'] != 'Other (please specify)']
new_dataframe = new_dataframe[new_dataframe['What are you most excited about as we move toward a more digitally connected future?'] != 'Other (please specify)']
new_dataframe = new_dataframe[new_dataframe['I consider myself:'] != 'Other (please specify)']
new_dataframe
```

```python
#new_dataframe['I consider myself:'] = new_dataframe['I consider myself:'] .astype(str) 
#Remap expertise question to 1-4
new_dataframe['I consider myself:'] = new_dataframe['I consider myself:'].replace('Ultra Nerd:  I build my own computers, run my own servers, code my own apps. I\x89\xdb\xaam basically Mr. Robot.',4)
new_dataframe['I consider myself:'] = new_dataframe['I consider myself:'].replace('Technically Savvy:   I know my way around a computer pretty well. When anyone in my family needs technical help, I\x89\xdb\xaam the one they call.',3)
new_dataframe['I consider myself:'] = new_dataframe['I consider myself:'].replace('Average User:   I know enough to get by.',2)
new_dataframe['I consider myself:'] = new_dataframe['I consider myself:'].replace('Luddite:  Technology scares me! I only use it when I have to.',1)

#Remap ownership question to 1/0
new_dataframe['WiFi Router:Check all the internet connected devices you currently own:']  = new_dataframe['WiFi Router:Check all the internet connected devices you currently own:'] .replace('WiFi Router',1)
new_dataframe['Laptop computer:Check all the internet connected devices you currently own:']  = new_dataframe['Laptop computer:Check all the internet connected devices you currently own:'] .replace('Laptop computer',1)
new_dataframe['Smart phone:Check all the internet connected devices you currently own:']  = new_dataframe['Smart phone:Check all the internet connected devices you currently own:'] .replace('Smart phone',1)
new_dataframe['Smart TV:Check all the internet connected devices you currently own:']  = new_dataframe['Smart TV:Check all the internet connected devices you currently own:'] .replace('Smart TV',1)
new_dataframe['Activity Tracker (ex: Fitbit or Apple Watch):Check all the internet connected devices you currently own:']  = new_dataframe['Activity Tracker (ex: Fitbit or Apple Watch):Check all the internet connected devices you currently own:'] .replace('Activity Tracker (ex: Fitbit or Apple Watch)',1)
new_dataframe['Smarthome Hub (ex. Amazon Echo, Google Alexa):Check all the internet connected devices you currently own:']  = new_dataframe['Smarthome Hub (ex. Amazon Echo, Google Alexa):Check all the internet connected devices you currently own:'] .replace('Smarthome Hub (ex. Amazon Echo, Google Alexa)',1)
new_dataframe['Car that connects to the internet:Check all the internet connected devices you currently own:']  = new_dataframe['Car that connects to the internet:Check all the internet connected devices you currently own:'] .replace('Car that connects to the internet',1)
new_dataframe['Smart Thermostat (ex: Nest):Check all the internet connected devices you currently own:']  = new_dataframe['Smart Thermostat (ex: Nest):Check all the internet connected devices you currently own:'] .replace('Smart Thermostat (ex: Nest)',1)
new_dataframe['Smart Appliance (ex. Coffeemaker, Refrigerator, Oven, Fridge):Check all the internet connected devices you currently own:']  = new_dataframe['Smart Appliance (ex. Coffeemaker, Refrigerator, Oven, Fridge):Check all the internet connected devices you currently own:'] .replace('Smart Appliance (ex. Coffeemaker, Refrigerator, Oven, Fridge)',1)
new_dataframe['Smart Door Locks (ex. Door locks for your home you can open via bluetooth):Check all the internet connected devices you currently own:']  = new_dataframe['Smart Door Locks (ex. Door locks for your home you can open via bluetooth):Check all the internet connected devices you currently own:'] .replace('Smart Door Locks (ex. Door locks for your home you can open via bluetooth)',1)
new_dataframe['Smart Lighting (ex. Connected lighting switches, dimmers, or bulbs):Check all the internet connected devices you currently own:']  = new_dataframe['Smart Lighting (ex. Connected lighting switches, dimmers, or bulbs):Check all the internet connected devices you currently own:'] .replace('Smart Lighting (ex. Connected lighting switches, dimmers, or bulbs)',1)

new_dataframe
```

```python
#Remap sentiment question to -2 to 2
new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:']  = new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:'] .replace('Scared as hell. The future where everything is connected has me scared senseless. We\x89\xdb\xaare all doomed!',-2)
new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:']  = new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:'] .replace('Cautiously optimistic. I\x89\xdb\xaam hopeful we\x89\xdb\xaare building a better world by becoming more connected in everything we do.',1)
new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:']  = new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:'] .replace('A little wary. All this being connected to the internet in every part of our lives makes me a little nervous. What\x89\xdb\xaas going to happen to our privacy?',-1)
new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:']  = new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:'] .replace('On the fence.  I\x89\xdb\xaam not sure about all this. I think I\x89\xdb\xaall wait and see.',0)
new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:']  = new_dataframe['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:'] .replace('Super excited! I can\x89\xdb\xaat wait for everything to be connected. My life will be so much better.',2)
new_dataframe
```

```python
#replace NaN with 0s
for name in filteredColumns:
    new_dataframe[name]  = new_dataframe[name].replace(np.nan,0)

new_dataframe
    
```

```python
#calculate number of owned devices
new_dataframe['Total Owned Devices'] = np.nan


for i,row in new_dataframe.iterrows():
    d1=row['WiFi Router:Check all the internet connected devices you currently own:'] 
    d2=row['Laptop computer:Check all the internet connected devices you currently own:'] 
    d3=row['Smart phone:Check all the internet connected devices you currently own:']  
    d4=row['Smart TV:Check all the internet connected devices you currently own:']  
    d5=row['Activity Tracker (ex: Fitbit or Apple Watch):Check all the internet connected devices you currently own:'] 
    d6=row['Smarthome Hub (ex. Amazon Echo, Google Alexa):Check all the internet connected devices you currently own:'] 
    d7=row['Car that connects to the internet:Check all the internet connected devices you currently own:']  
    d8=row['Smart Thermostat (ex: Nest):Check all the internet connected devices you currently own:']  
    d9=row['Smart Appliance (ex. Coffeemaker, Refrigerator, Oven, Fridge):Check all the internet connected devices you currently own:']
    d10=row['Smart Door Locks (ex. Door locks for your home you can open via bluetooth):Check all the internet connected devices you currently own:']
    d11=row['Smart Lighting (ex. Connected lighting switches, dimmers, or bulbs):Check all the internet connected devices you currently own:']      
    
    new_dataframe.at[i, 'Total Owned Devices'] = d1+d2+d2+d4+d5+d6+d7+d8+d9+d10+d11
new_dataframe
```

```python
#Adding Country Average Devices
countries=new_dataframe.Country.unique()
print(countries)
```

### Remove Countries with less than 1000 records

```python
s = new_dataframe['Country'].value_counts()
countries = s.index[s >= 1000]
print(s)
new_dataframe = new_dataframe[new_dataframe.isin(s.index[s >= 1000]).values]

```

### Calculate Average Devices owned by country

```python
# average count for each country
new_dataframe['Country Average Devices'] = np.nan

print countries
for country in countries:
    a = new_dataframe[new_dataframe['Country'] == country]
    totalForCountry = sum(a['Total Owned Devices'])
    num_rows = a.shape[0]
    averageForCountry = ((totalForCountry / (num_rows*1.0)))
    print(str(averageForCountry) + ": " + country)
    
    new_dataframe.loc[new_dataframe.Country == country, ['Country Average Devices']] = averageForCountry

print new_dataframe
```

### Calculate Average Sentiment per country

```python
# average count for each country
new_dataframe['Country Average Tech Sentiment'] = np.nan
totalForCountry=0
for country in countries:
    a = new_dataframe[new_dataframe['Country'] == country] 
    totalForCountry = a['Thinking about a future in which so much of your world is connected to the internet leaves you feeling:'].sum()
    num_rows = a.shape[0]
    print(num_rows)
    averageForCountry = (totalForCountry / (num_rows*1.0))
    print(str((totalForCountry / num_rows)) + " " + str(averageForCountry) + ": " + country + " Tech Proficiency")
    
    new_dataframe.loc[new_dataframe.Country == country, ['Country Average Tech Sentiment']] = averageForCountry

print new_dataframe
```
