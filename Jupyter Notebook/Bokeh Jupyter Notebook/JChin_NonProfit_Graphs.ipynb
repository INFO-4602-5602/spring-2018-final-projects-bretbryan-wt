```python
scatterdata_q1 = pd.read_csv("BTC Anlys 20180428 satisfaction only.csv", sep=',')
r2data = pd.read_csv("BTC Anlys 20180428 satisfaction r value only version 02.csv", sep='\t')
```

```python
question_group=r2data["Question Group"].unique()
question_group
```

```python
#custom color scheme
def Set2_8(self):
         return ["#e41a1c", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00"]
    
def color_scheme_categories(self):
         return ["#b3e2cd", "#fdcdac", "#cbd5e8", "#f4cae4", "#e6f5c9"]
    
```

```python
#Setting up Bokeh
from bokeh.io import output_notebook,output_file
output_notebook()
```

```python
from bokeh.io import show, output_file
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, TapTool
from bokeh.models import HoverTool, Range1d
from bokeh.models import LinearInterpolator, CategoricalColorMapper
from bokeh.palettes import inferno
from bokeh.layouts import gridplot
from bokeh.models.callbacks import CustomJS
from ipywidgets import interact
from bokeh.io import push_notebook, show, output_notebook
from bokeh.embed import file_html
from bokeh.resources import CDN
from bokeh.models import Legend


r2source = ColumnDataSource(r2data)


def generateBoxPlot(correlations):  
    if correlations is "Top3Corr":
        filteredScatter = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q37"]
        filteredScatter2 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q21"]
        filteredScatter3 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q16"]
        
    elif correlations is "Bottom3Corr":
        filteredScatter = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q09"]
        filteredScatter2 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q08"]
        filteredScatter3 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q03"]
        
    elif correlations is "Top3_Self_Reported":
        filteredScatter = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q05"]
        filteredScatter2 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q04"]
        filteredScatter3 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q06"]
        
    elif correlations is "Top3_Education":
        filteredScatter = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q10"]
        filteredScatter2 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q12"]
        filteredScatter3 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q13"]   
        
    elif correlations is "Top3_Staff":
        filteredScatter = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q37"]
        filteredScatter2 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q21"]
        filteredScatter3 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q16"]
        
    elif correlations is "Top3_Safety":
        filteredScatter = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q23"]
        filteredScatter2 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q24"]
        filteredScatter3 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q25"]
        
    elif correlations is "Top3_Impact":
        filteredScatter = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q33"]
        filteredScatter2 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q30"]
        filteredScatter3 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q29"]
        
    elif correlations is "Top3_Support":
        filteredScatter = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q35"]
        filteredScatter2 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q39"]
        filteredScatter3 = scatterdata_q1[scatterdata_q1["QuestionNumber"] == "Q38"]
    elif correlations is "All_Staff":
        GenerateStaffPlots()
        return
        
    
    source1 = ColumnDataSource(filteredScatter) 
    source2 = ColumnDataSource(filteredScatter2)     
    source3 = ColumnDataSource(filteredScatter3) 

    questionStr = filteredScatter.iloc[0]['QuestionString']
    questionStr2 = filteredScatter2.iloc[0]['QuestionString']
    questionStr3 = filteredScatter3.iloc[0]['QuestionString']
        
        
    
    color_mapper = CategoricalColorMapper(
        factors = question_group,
        palette = color_scheme_categories(6)
    )
    

    hover = HoverTool(tooltips = [
        ("Question:","@QuestionString"),
        ("R Satisfaction:" ,"@R_Satisfaction"),
        #("Rank:" ,"@Rank"),
        #("Question Group:" ,"@Question Group"),
    ], show_arrow=False)
    
    hover2 = HoverTool(tooltips = [
        ("Country","@Country"),
        ("Satisfaction Response:","@SatisfactionResponse"),
        ("Average Question Response:" ,"@QuestionResponse"),
        ("Count:" ,"@Count"),
    ], show_arrow=False)


    #interpolating from min max values
    size_mapper = LinearInterpolator(
        x = [filteredScatter['Count'].min(), filteredScatter['Count'].max()],
        y=[5,20]
    )
    
    size_mapper2 = LinearInterpolator(
        x = [filteredScatter2['Count'].min(), filteredScatter2['Count'].max()],
        y=[5,20]
    )
    
    size_mapper3 = LinearInterpolator(
        x = [filteredScatter3['Count'].min(), filteredScatter3['Count'].max()],
        y=[5,20]
    )
    
    color_mapper2 = CategoricalColorMapper(
        factors = country_group,
            palette = Set2_8(5)
    )
    
    
    plot1 = figure(
            title=str("R-values correlating Survey Questions to Satisfaction Response of NonProfit Program"),
            x_axis_label='R correlation (+1) positively correlated (0) no correlation', y_axis_label='Correlation Ranking among questions',
            tools=[hover,"tap"],
            x_range = Range1d(0,1)
    )
    plot1.title.text_font_size = '10pt'
    

    ro=plot1.circle(
            x='R_Satisfaction', y='Rank', 
            #size= {'field': 'Count','transform':size_mapper},
            color= {'field': 'Question Group', 'transform':color_mapper},     
            source=r2source,
            size=10,
            legend="Question Group"
    )
    
    
    scatterplot_q1 = figure(
            title=str(questionStr),
            x_axis_label='Question Average  Response', y_axis_label='Satisfaction Response',
            tools=[hover2,"tap"],
            x_range = Range1d(-2,2),
            y_range = Range1d(-2,2),
            name="scatter"
    )
    
    rx=scatterplot_q1.circle(
            x='QuestionResponse', y='SatisfactionResponse', 
            size= {'field': 'Count','transform':size_mapper},
            color= {'field': 'Country', 'transform':color_mapper2},     
            source=source1,
            legend="Country"
    )
    
    q1_glyph = rx.glyph
    q1_glyph.fill_alpha = .4
    
    
    scatterplot_q2 = figure(
            title=str(questionStr2),
            x_axis_label='Question Average Response', y_axis_label='Satisfaction Response',
            tools=[hover2,"tap"],
            x_range = Range1d(-2,2),
            y_range = Range1d(-2,2),
            name="scatter"
    )
    
    rx2=scatterplot_q2.circle(
            x='QuestionResponse', y='SatisfactionResponse', 
            size= {'field': 'Count','transform':size_mapper2},
            color= {'field': 'Country', 'transform':color_mapper2},     
            source=source2,
            legend="Country"
    )
    
    q2_glyph = rx2.glyph
    q2_glyph.fill_alpha = .4
    
    
    scatterplot_q3 = figure(
            title=str(questionStr3),
            x_axis_label='Question Average Response', y_axis_label='Satisfaction Response',
            tools=[hover2,"tap"],
            x_range = Range1d(-2,2),
            y_range = Range1d(-2,2),
            name="scatter"
    )
    
    rx3=scatterplot_q3.circle(
            x='QuestionResponse', y='SatisfactionResponse', 
            size= {'field': 'Count','transform':size_mapper3},
            color= {'field': 'Country', 'transform':color_mapper2},     
            source=source3,
            legend="Country"
    )
    q3_glyph = rx3.glyph
    q3_glyph.fill_alpha = .4
    
    plot1.legend.location = "bottom_right"
    
    scatterplot_q1.legend.location = "top_left"
    
    scatterplot_q2.legend.location = "top_left"
    
    scatterplot_q3.legend.location = "top_left"
    
    
    plots = gridplot([plot1,scatterplot_q1,scatterplot_q2,scatterplot_q3], ncols=2, plot_width=600, plot_height=600)
    show(plots)#, notebook_handle=True)
    
    output_file("JChin_NonProfit_Graphs.html")
```
