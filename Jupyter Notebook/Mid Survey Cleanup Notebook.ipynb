## Filtering and Cleaning Mid Survey Data

```python
mid_service_dataframe = pd.read_csv(mid_service_evaluation)
mid_service_dataframe
```

```python
list(mid_service_dataframe)
```

## Filter columns to ones of interest

```python
#Columns kept
filteredColumns =['Program Country:', 
                  'Program Start Year:',
                  'I would describe myself as:Resourceful',
                  'I would describe myself as:Patient',
                  'I would describe myself as:Culturally Competent',
                  'I would describe myself as:Adaptable',
                  'I would describe myself as:Emotionally Intelligent',
                  'I would describe myself as:Resilient',
                  'Professional Learning Communities (PLCs) are useful to me',
                  'The Five Step Lesson Plan Template is useful to me',
                  'The Data Analysis Tool is useful to me',
                  'The Professional Development Modules (Scholarly Articles) are useful to me',
                  'WorldTeachNet is useful to me',
                  'The Teacher Quality Coordinator is useful to me',
                  'The Director of Education Office Hours are useful to me',
                  'I am satisfied in the Educational Support I have access to from WorldTeach',
                  'The staff member:Is accessible and eager to help.',
                  'The staff member:Responds promptly to requests for help.',
                  'The staff member:Anticipates needs of volunteers.',
                  'The staff member:Follows through on administrative, site, or housing issues.',
                  'The staff member:Keeps volunteers informed of changes and relevant information.',
                  'Has the Field Director conducted a site visit with you yet?',
                  'The site visit was useful to me.',
                  'The Field Director made me feel comfortable and supported.',
                  'The Field Director gave me useful and professional suggestions.',
                  'I feel safe in my home.',
                  'I feel safe in my community.',
                  'I feel integrated into my community, increasing my sense of safety.',
                  'I feel welcome as a foreigner in my community.',
                  'I have friends, host family members, colleagues, or community members who would help protect me in an emergency situation.',
                  'The coverage of health and safety issues at Orientation was sufficient.',
                  'I have an understanding of what to do in various emergency situations, as laid out in the Emergency Action Plan.',
                  'I always inform the Field Director of my travel plans when I go off-site.',
                  'I generally feel safe in my country.']
new_mid_service_dataframe = mid_service_dataframe[filteredColumns]
new_mid_service_dataframe
```

```python
### Remove Null Values from questions and country column
new_mid_service_dataframe = new_mid_service_dataframe[new_mid_service_dataframe['I would describe myself as:Resourceful'].notnull()]
new_mid_service_dataframe 
```

```python
#new_dataframe['I consider myself:'] = new_dataframe['I consider myself:'] .astype(str) 
new_mid_service_dataframe=new_mid_service_dataframe.replace('Agree',1)
new_mid_service_dataframe=new_mid_service_dataframe.replace('Disagree',-1)
new_mid_service_dataframe=new_mid_service_dataframe.replace('Strongly Agree',2)
new_mid_service_dataframe=new_mid_service_dataframe.replace('Strongly Disagree',-2)
new_mid_service_dataframe=new_mid_service_dataframe.replace('Neutral',0)

new_mid_service_dataframe=new_mid_service_dataframe.replace('Yes',1)

new_mid_service_dataframe=new_mid_service_dataframe.replace('No',-1)
new_mid_service_dataframe 
```

```python
#replace NaN with 0s
for name in filteredColumns:
    new_mid_service_dataframe[name]  = new_mid_service_dataframe[name].replace(np.nan,0)

new_mid_service_dataframe
    
```

```python
#Adding Country Average Devices
countries=new_mid_service_dataframe['Program Country:'].unique()
print(countries)
```

```python
s = new_mid_service_dataframe['Program Country:'].value_counts()
countries = s.index[s >= 1]
print(s)
```

### Calculate Averages by country

```python
column_list = list(new_mid_service_dataframe)
column_list = column_list[2:len(column_list)]  #ignore first two columns (country and date)

# average responses for each country
for column in column_list:
    print column
    new_mid_service_dataframe['Average:'+column] = np.nan
    for country in countries:
        a = new_mid_service_dataframe[new_mid_service_dataframe['Program Country:'] == country]
        totalForCountry = sum(a[column])
        num_rows = a.shape[0]
        averageForCountry = ((totalForCountry / (num_rows*1.0)))
        print(str(averageForCountry) + ": " + country)
        new_mid_service_dataframe.loc[new_mid_service_dataframe['Program Country:'] == country, ['Average:'+column]] = averageForCountry

print new_mid_service_dataframe
```

```python
#Output results to csv
new_mid_service_dataframe.to_csv("avg_mid_service_surveys.csv", sep='\t')
```
