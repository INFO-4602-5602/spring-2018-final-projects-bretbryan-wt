## Filter columns to ones of interest

```python
#Columns kept
filteredColumns =['Program Country:', 
                  'Program Start Year:',
                  'I would describe myself as: Resourceful',
                  'I would describe myself as: Patient',
                  'I would describe myself as: Culturally Competent',
                  'I would describe myself as: Adaptable',
                  'I would describe myself as: Emotionally Intelligent',
                  'I would describe myself as: Resilient',
                  'I am satisfied with the support I received from WorldTeach prior to departure.',
                  'I am\xc2\xa0satisfied with the support I received during Orientation',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Addressed group needs well\r\n',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Encouraged group cohesiveness well\r\n',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Was sensitive to individual concerns\r\n',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Was accessible for individual questions\r\n',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Encouraged/was open to feedback\r\n',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Was knowledgeable about country/culture\r\n',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Was knowledgeable about training topics\r\n',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Gave clear and effective presentations',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Was well-organized\r\n',
                  'Combined: Please rate your field staff\xc2\xa0on the following: Was approachable']
new_orientation_dataframe = orientation_dataframe[filteredColumns]
new_orientation_dataframe
```

```python
### Remove Null Values from questions and country column
new_orientation_dataframe = new_orientation_dataframe[new_orientation_dataframe['I would describe myself as: Resourceful'].notnull()]
new_orientation_dataframe 
```

```python
#new_dataframe['I consider myself:'] = new_dataframe['I consider myself:'] .astype(str) 
new_orientation_dataframe=new_orientation_dataframe.replace('Agree',1)
new_orientation_dataframe=new_orientation_dataframe.replace('Disagree',-1)
new_orientation_dataframe=new_orientation_dataframe.replace('Strongly Agree',2)
new_orientation_dataframe=new_orientation_dataframe.replace('Strongly Disagree',-2)
new_orientation_dataframe=new_orientation_dataframe.replace('Neutral',0)
new_orientation_dataframe=new_orientation_dataframe.replace('DisagreeAgree',0)

new_orientation_dataframe=new_orientation_dataframe.replace('AgreeStrongly Agree',1.5)
new_orientation_dataframe=new_orientation_dataframe.replace('Yes',1)

new_orientation_dataframe=new_orientation_dataframe.replace('No',-1)
new_orientation_dataframe 
```

```python
#replace NaN with 0s
for name in filteredColumns:
    new_orientation_dataframe[name]  = new_orientation_dataframe[name].replace(np.nan,0)

new_orientation_dataframe
    
```

```python
#Adding Country Average Devices
countries=new_orientation_dataframe['Program Country:'].unique()
print(countries)
```

### Calculate Averages by country

```python
column_list = list(new_orientation_dataframe)
column_list = column_list[2:len(column_list)]  #ignore first two columns (country and date)

# average responses for each country
for column in column_list:
    print column
    new_orientation_dataframe['Average:'+column] = np.nan
    for country in countries:
        a = new_orientation_dataframe[new_orientation_dataframe['Program Country:'] == country]
        totalForCountry = sum(a[column])
        num_rows = a.shape[0]
        averageForCountry = ((totalForCountry / (num_rows*1.0)))
        print(str(averageForCountry) + ": " + country)
        new_orientation_dataframe.loc[new_orientation_dataframe['Program Country:'] == country, ['Average:'+column]] = averageForCountry

print new_orientation_dataframe
```
